```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_2020_cleaned.csv')
```

```python
df.head(5)
```

```python
Low = df[
    (df["BMI"] < 20) & 
    (df["GenHealth"].isin(['Very good', 'Excellent'])) &
    (df["Smoking"] == 'No')&
    (df["AlcoholDrinking"] == 'No')&
    (df["PhysicalActivity"] == 'Yes')&
    (df["DiffWalking"] == 'No')&
    (df["HeartDisease"] == 'No')]
Low.shape
```

```python
High = df[
    (df["BMI"] >= 30) & 
    (df["GenHealth"].isin(['Poor']))&
    (df["HeartDisease"] == 'Yes')]
High.shape
```

```python
Mid = df[
    (df["BMI"].between(18.5, 24.9, inclusive='neither'))& 
    (df["GenHealth"].isin(['Fair','Good']))]
Mid.shape
```

```python
Low["HeartDisease"] = 'Low'
Mid["HeartDisease"] = 'Mid'
High["HeartDisease"] = 'High'
```

```python
df = pd.concat([Low, Mid,High])
df.head(5)
```

```python
df.to_csv('output.csv', index=False)
```
